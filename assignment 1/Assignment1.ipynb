```python
# prompt: counting unique values of data, format of each column, converting variable data type (e.g. from long to short, vice versa)

# h) counting unique values of data, format of each column, converting variable
# data type (e.g. from long to short, vice versa)

# Counting unique values
for col in data.columns:
  print(f"Unique values in column '{col}': {data[col].nunique()}")

# Format of each column (data type)
print("\nData types of each column:")
print(data.dtypes)

# Converting variable data type (example: converting 'Age' from int to float)
# Note that this changes the data; check if this is appropriate for the task
data['Age'] = data['Age'].astype(float)

# Verify the type change
print("\nData types after conversion:")
print(data.dtypes)

# Example of converting a numerical column to a categorical type with specific bins
# This is just one example; adapt binning strategy as needed.
data['AgeGroup'] = pd.cut(data['Age'], bins=[0, 20, 40, 60, 80, 100], labels=['0-20', '20-40', '40-60', '60-80', '80-100'])
print(data.head())
```
