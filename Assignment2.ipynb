```python
import pandas as pd
import numpy as py
```

```python
data=pd.read_csv("/content/sample_data/Heart (1).csv")
```

```python
data.head()
#displaying the data
```

```python
#now cleaning of data.
#1 finding missing values
#2 replace it with mean

data.isnull().sum()
```

```python
data['Ca'].fillna(data['Ca'].mean(), inplace=True)
data['Thal'].fillna(data['Thal'].mode(), inplace=True)
```

```python
data.isnull().sum()
```

Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
#Bar Plot (Count of Chest Pain Types)
plt.figure(figsize=(8,5))
sns.countplot(x=data['ChestPain'])
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")
plt.title("Bar Plot: Chest Pain Type Distribution")
plt.show()
```

```python
data.columns
```

```python
#scatter graph

plt.figure(figsize=(8,5))
sns.scatterplot(x=data['Age'], y=data['MaxHR'], alpha=0.7)
plt.xlabel("Age")
plt.ylabel("Maximum Heart Rate (MaxHR)")
plt.title("Scatter Plot: Age vs MaxHR")
plt.show()
```

```python
#Box Plot (Cholesterol Levels by Heart Disease)

plt.figure(figsize=(8,5))
sns.boxplot(x=data['AHD'], y=data['Chol'])
plt.xlabel("Heart Disease (AHD)")
plt.ylabel("Cholesterol Level")
plt.title("Box Plot: Cholesterol Levels by Heart Disease")
plt.show()
```

```python
#Pie Chart (Sex Distribution)
plt.figure(figsize=(5,5))
data['Sex'].value_counts().plot.pie(autopct="%1.1f%%", colors=["lightblue", "pink"])
plt.ylabel("")
plt.title("Pie Chart: Gender Distribution")
plt.show()
```

```python
#Line Chart (Age vs Resting Blood Pressure)
plt.figure(figsize=(8,5))
sns.lineplot(x=data['Age'], y=data['RestBP'])
plt.xlabel("Age")
plt.ylabel("Resting Blood Pressure (RestBP)")
plt.title("Line Chart: Age vs Resting BP")
plt.show()

```
